**Ans:** The variable name should not start from the starting number. The name of the variable should be given according to its rules

Rules for Python variables:

1. A variable name must start with a letter or the underscore character.
2. A variable name cannot start with a number.
3. A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ ).
4. Variable names are case-sensitive (age, Age and AGE are three different variables)

```python
integer = 10
float_ = 56.5
string = 56
```

```python
# Examples:
print('int(56.5) -> ',int(float_)) # int() function converts given input to int
print('float(10) -> ',float(integer)) # float() function converts given input to float
print('str(56) -> ',str(string)) # str() function converts given input to string
```

```python
print('type(integer) ',type(integer))
print('type(float) ',type(float_))
print('type(string) ',type(string))
```
